```python
import mbuild as mb
import warnings
import foyer
from mbuild.lib.recipes import Polymer
warnings.filterwarnings('ignore')
```

```python
class y6_chunk(mb.Compound):
    def __init__(self):
        super(y6_chunk, self).__init__()
        self.add(mb.load('/Users/gwen/cmelab/molecules/y6-4bto.mol2'))
        carbon1 = list(self.particles_by_name('C'))[22]
        carbon2 = list(self.particles_by_name('C'))[23]
        self.add(mb.Port(anchor=carbon1, orientation=[0.01, -1, 1], separation=0.014), "y61")
        self.add(mb.Port(anchor=carbon2, orientation=[0.01, -1, -1], separation=0.014), "y62")
y6_chunk = y6_chunk()
y6_chunk.visualize(show_ports=True)
```

```python
class bto_tail1(mb.Compound):
    def __init__(self):
        super(bto_tail1, self).__init__()
        self.add(mb.load('/Users/gwen/cmelab/molecules/bto_tail.mol2'))
        carbon = list(self.particles_by_name('C'))[0]
        self.add(mb.Port(anchor=carbon, orientation=[-0.04, 0.025, 0], separation=0.14), "tail")
bto_tail1 = bto_tail1()
bto_tail1.visualize(show_ports=True)
```

```python
class bto_tail2(mb.Compound):
    def __init__(self):
        super(bto_tail2, self).__init__()
        self.add(mb.load('/Users/gwen/cmelab/molecules/bto_tail.mol2'))
        carbon = list(self.particles_by_name('C'))[0]
        self.add(mb.Port(anchor=carbon, orientation=[-0.04, 0.025, 0], separation=0.14), "tail")
bto_tail2 = bto_tail2()
bto_tail2.visualize(show_ports=True)
```

```python
parent = mb.Compound()
parent.add((y6_chunk, bto_tail2, bto_tail1))
mb.force_overlap(move_this=bto_tail2,
                 from_positions=bto_tail2['tail'],
                 to_positions=y6_chunk['y62'])
mb.force_overlap(move_this=bto_tail1,
                 from_positions=bto_tail1['tail'],
                to_positions=y6_chunk['y61'])
#parent.energy_minimize()
parent.visualize(show_ports=True)
```
