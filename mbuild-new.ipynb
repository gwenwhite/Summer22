```python
import mbuild as mb
import warnings
warnings.filterwarnings('ignore')
```

```python
class bto_fring(mb.Compound):
    def __init__(self):
        super(bto_fring, self).__init__()
        self.add(mb.load('/Users/gwen/cmelab/molecules/y6-lastring3.sdf'))
        carbon = list(self.particles_by_name('C'))[1]
        self.add(mb.Port(anchor=carbon, orientation=[0, 1, 0], separation=0.01), "fring")
bto_fring = bto_fring()
bto_fring.visualize(show_ports=True)
```

```python
class bto_sring(mb.Compound):
    def __init__(self):
        super(bto_sring, self).__init__()
        self.add(mb.load('/Users/gwen/cmelab/molecules/y6_sring.mol2'))
        carbon = list(self.particles_by_name('C'))[8]
        carbon1 = list(self.particles_by_name('C'))[7]
        hydrogen = list(self.particles_by_name('H'))[38]
        self.add(mb.Port(anchor=carbon, orientation=[-1, 0.5, 1], separation=0.01), "sringc1")
        self.add(mb.Port(anchor=carbon1, orientation=[-1, 0.5, 1], separation=0.01), "sringc2")
        self.add(mb.Port(anchor=hydrogen, orientation=[0, 1, 1], separation=0.01), "sringh")
bto_sring=bto_sring()
bto_sring.visualize(show_ports=True)
```

```python
class bto_top(mb.Compound):
    def __init__(self):
        super(bto_top, self).__init__()
        self.add(mb.load('/Users/gwen/cmelab/molecules/y6-5ring2.mol2'))
        hydrogen = list(self.particles_by_name('H'))[0]
        hydrogen1 = list(self.particles_by_name('H'))[1]
        hydrogen2 = list(self.particles_by_name('H'))[2]
        hydrogen3 = list(self.particles_by_name('H'))[3]
        self.add(mb.Port(anchor=hydrogen, orientation=[0, -1, 0], separation=0.03), "top1")
        self.add(mb.Port(anchor=hydrogen1, orientation=[0, -1, 0], separation=0.02), "top2")
        self.add(mb.Port(anchor=hydrogen2, orientation=[0, 1, 0], separation=0.03), "top3")
        self.add(mb.Port(anchor=hydrogen3, orientation=[0, 1, 0], separation=0.03), "top4")
bto_top= bto_top()
bto_top.visualize(show_ports=True)
```

```python
bto_sring.translate([0, 1, 0])
bto_fring.translate([0, -1, 0])
parent = mb.Compound()
parent.add((bto_top, bto_sring, bto_fring))
parent.visualize(show_ports=True)
```

```python
mb.force_overlap(move_this=bto_top,
                 from_positions=bto_top['top3'],
                 to_positions=bto_sring['sringc2'])
mb.force_overlap(move_this=bto_top,
                 from_positions=bto_top['top4'],
                 to_positions=bto_sring['sringc1'])
mb.force_overlap(move_this=bto_fring,
                 from_positions=bto_fring['fring'],
                 to_positions=bto_sring['sringh'])
parent.visualize(show_ports=True)
```
